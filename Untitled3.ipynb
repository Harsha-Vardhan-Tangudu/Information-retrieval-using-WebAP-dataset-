```python
import re
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from unidecode import unidecode  # To handle accents
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')  # Necessary for WordNetLemmatizer

def preprocess(text, use_lemmatization=False):
    # Normalize text to lower case, remove accents, hyphens, and non-word characters
    text = unidecode(text.lower())
    text = re.sub(r'-', ' ', text)
    text = re.sub(r'\W+', ' ', text)
    tokens = nltk.word_tokenize(text)
    
    # Use either stemming or lemmatization
    if use_lemmatization:
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stopwords.words('english')]
    else:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(token) for token in tokens if token not in stopwords.words('english')]
    
    return tokens

# Generator function to parse TRECTEXT and yield documents one at a time
def parse_trectext(file_path):
    current_doc_id = None
    capture_text = False
    text_buffer = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if '<DOCNO>' in line:
                current_doc_id = re.search('<DOCNO>(.*?)</DOCNO>', line).group(1).strip()
            elif '<TEXT>' in line:
                capture_text = True
            elif '</TEXT>' in line:
                capture_text = False
                if current_doc_id and text_buffer:
                    processed_text = preprocess(' '.join(text_buffer))
                    yield (current_doc_id, processed_text)
                    text_buffer = []  # Reset buffer after processing
            elif capture_text:
                text_buffer.append(line.strip())

# Function to process the first 10 documents
def process_first_10_documents(file_path):
    doc_count = 0
    for doc_id, processed_text in parse_trectext(file_path):
        doc_count += 1
        print(f"Document ID: {doc_id}")
        print("Processed Text:", processed_text)
        print("="*80)  # Separator for readability

        if doc_count == 10:
            break  # Stop after processing the first 10 documents

# Example usage
file_path = 'path_to_your_reduced_file'  # Replace with your actual file path
process_first_10_documents(file_path)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import defaultdict
from unidecode import unidecode

# Ensure necessary NLTK resources are available
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Initialize stop words and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Function to parse the custom document format
def parse_custom_corpus(file_lines):
    documents = {}
    doc_id = None
    doc_text = []
    
    for line in file_lines:
        if "<DOCNO>" in line:
            doc_id = re.search(r'<DOCNO>(.*?)</DOCNO>', line).group(1)
        elif "<TEXT>" in line:
            doc_text = []  # Reset for new document text
        elif "<SENTENCE>" in line:
            sentence = re.search(r'<SENTENCE>(.*?)</SENTENCE>', line).group(1)
            doc_text.append(sentence)
        elif "</DOC>" in line and doc_id:
            documents[doc_id] = ' '.join(doc_text)
            doc_id = None
    
    return documents

# Function for normalizing text (lowercasing, removing accents)
def normalize(text):
    # Lowercase and remove accents
    return unidecode(text.lower())

# Function for preprocessing text (tokenization, stopword removal, lemmatization)
def preprocess(text):
    text = normalize(text)
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return lemmatized_tokens

# Function to build the inverted index with term positions
def build_inverted_index(docs):
    inverted_index = defaultdict(lambda: defaultdict(list))
    for doc_id, tokens in docs.items():
        for pos, token in enumerate(tokens):
            inverted_index[token][doc_id].append(pos)
    return inverted_index

# Boolean query processing (AND, OR, NOT)
def boolean_query(query, inverted_index, operation="AND"):
    if operation == "AND":
        result = set(inverted_index.get(query[0], {}))
        for term in query[1:]:
            result &= set(inverted_index.get(term, {}))
    elif operation == "OR":
        result = set()
        for term in query:
            result |= set(inverted_index.get(term, {}))
    elif operation == "NOT":
        all_docs = set(inverted_index.keys())
        result = all_docs - set(inverted_index.get(query[0], {}))
    
    return result

# Proximity query processing
def proximity_query(term1, term2, max_distance, inverted_index):
    result = set()
    term1_docs = inverted_index.get(term1, {})
    term2_docs = inverted_index.get(term2, {})
    
    common_docs = set(term1_docs.keys()).intersection(term2_docs.keys())
    
    for doc_id in common_docs:
        term1_positions = term1_docs[doc_id]
        term2_positions = term2_docs[doc_id]
        for pos1 in term1_positions:
            for pos2 in term2_positions:
                if abs(pos1 - pos2) <= max_distance:
                    result.add(doc_id)
                    break
    return result

# Phrase query processing
def phrase_query(terms, inverted_index):
    if len(terms) < 2:
        return inverted_index.get(terms[0], {})

    result = inverted_index.get(terms[0], {})
    
    for i in range(1, len(terms)):
        term_result = inverted_index.get(terms[i], {})
        new_result = {}
        
        for doc_id, positions in result.items():
            if doc_id in term_result:
                for pos in positions:
                    if pos + 1 in term_result[doc_id]:
                        if doc_id not in new_result:
                            new_result[doc_id] = []
                        new_result[doc_id].append(pos + 1)
        
        result = new_result
    
    return set(result.keys())

# Load the dataset
file_path = 'path_to_your_reduced_file'  # Replace with your actual file path
with open(file_path, 'r', encoding='utf-8') as f:
    file_lines = f.readlines()

# Parse the custom corpus from the provided file lines
parsed_documents = parse_custom_corpus(file_lines)

# Preprocess each document's text
preprocessed_documents = {doc_id: preprocess(text) for doc_id, text in parsed_documents.items()}

# Build the inverted index with positions
inverted_index = build_inverted_index(preprocessed_documents)

# Displaying some terms and their corresponding document IDs for testing
print("Sample of terms and their document IDs:")
for term, doc_ids in list(inverted_index.items())[:10]:  # Display only the first 10 terms for testing
    print(f"Term: {term}, Document IDs: {list(doc_ids.keys())}")

# Example usage:
# Boolean Query (AND)
and_query_result = boolean_query(["august", "member"], inverted_index, operation="AND")
print("Boolean AND query result:", and_query_result)

# Proximity Query (within 3 words)
proximity_result = proximity_query("august", "member", 3, inverted_index)
print("Proximity query result:", proximity_result)

# Phrase Query
phrase_result = phrase_query(["august", "member"], inverted_index)
print("Phrase query result:", phrase_result)
```

```python
import numpy as np
from scipy.sparse import lil_matrix, csr_matrix

# Create a sparse term-document incidence matrix from the inverted index
def build_sparse_term_document_matrix(inverted_index, doc_ids):
    terms = sorted(inverted_index.keys())  # Sort terms alphabetically for consistent matrix order
    num_docs = len(doc_ids)
    num_terms = len(terms)

    # Initialize the incidence matrix using sparse matrix (LIL format is good for constructing the matrix)
    incidence_matrix = lil_matrix((num_terms, num_docs), dtype=int)

    # Create a mapping of document IDs to index for columns
    doc_id_to_index = {doc_id: i for i, doc_id in enumerate(sorted(doc_ids))}

    # Populate the incidence matrix
    for term_idx, term in enumerate(terms):
        for doc_id in inverted_index[term]:
            col_idx = doc_id_to_index[doc_id]
            incidence_matrix[term_idx, col_idx] = 1  # Set to 1 where term appears in document

    # Convert the matrix to CSR format for efficient operations
    return terms, incidence_matrix.tocsr()

# Get all document IDs
all_doc_ids = sorted(parsed_documents.keys())

# Build and print the sparse term-document incidence matrix
terms, sparse_term_document_matrix = build_sparse_term_document_matrix(inverted_index, all_doc_ids)

# Display basic information about the sparse matrix
print(f"Sparse Matrix Shape: {sparse_term_document_matrix.shape}")
print(f"Number of non-zero entries: {sparse_term_document_matrix.nnz}")

# Display the first few rows of the sparse matrix in dense format for inspection
print("\nFirst few rows of the sparse matrix (in dense format for viewing):")
print(sparse_term_document_matrix[:10, :10].todense())  # Display a small portion
```

```python
import math
from collections import defaultdict

# Function to add skip pointers to the inverted index
def add_skip_pointers(postings_list):
    """
    Adds skip pointers to a postings list.
    A skip pointer allows us to skip ahead in the list, 
    speeding up the intersection process.
    """
    skip_distance = int(math.sqrt(len(postings_list)))  # Using sqrt(N) heuristic for placing skips
    if skip_distance > 1:
        for i in range(0, len(postings_list), skip_distance):
            if i + skip_distance < len(postings_list):
                postings_list[i] = (postings_list[i], i + skip_distance)  # (docID, skip to index)
            else:
                postings_list[i] = (postings_list[i], None)  # No skip pointer at the end
    else:
        postings_list = [(doc, None) for doc in postings_list]  # No skip pointer needed for small lists
    return postings_list

# Function to build the inverted index with skip pointers
def build_inverted_index_with_skips(docs):
    inverted_index = defaultdict(list)
    for doc_id, tokens in docs.items():
        for token in tokens:
            inverted_index[token].append(doc_id)

    # Add skip pointers to each postings list
    for term, postings_list in inverted_index.items():
        inverted_index[term] = add_skip_pointers(postings_list)

    return inverted_index

# Function to perform intersection using skip pointers
def intersect_with_skip_pointers(list1, list2):
    """
    Intersects two posting lists with skip pointers.
    Posting list format: [(docID, skip_index), ...]
    """
    i, j = 0, 0
    result = []

    while i < len(list1) and j < len(list2):
        doc1, skip1 = list1[i]
        doc2, skip2 = list2[j]

        if doc1 == doc2:
            result.append(doc1)
            i += 1
            j += 1
        elif doc1 < doc2:
            if skip1 is not None and list1[skip1][0] <= doc2:
                i = skip1  # Skip ahead
            else:
                i += 1
        else:
            if skip2 is not None and list2[skip2][0] <= doc1:
                j = skip2  # Skip ahead
            else:
                j += 1

    return result

# Example dataset
preprocessed_documents = process_first_10_documents(file_path)

# Build the inverted index with skip pointers
inverted_index = build_inverted_index_with_skips(preprocessed_documents)

# Example postings lists for two terms
postings_august = inverted_index['august']  # e.g. [(1, None), (2, None)]
postings_member = inverted_index['member']  # e.g. [(1, None), (3, None)]

# Intersect the postings lists using skip pointers
result_with_skips = intersect_with_skip_pointers(postings_august, postings_member)

print(f"Intersection using skip pointers: {result_with_skips}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import defaultdict

# Step 1: Soundex Algorithm for phonetic matching
def soundex(word):
    word = word.upper()
    soundex_mapping = {'BFPV': '1', 'CGJKQSXZ': '2', 'DT': '3', 'L': '4', 'MN': '5', 'R': '6'}
    soundex_code = word[0]  # First letter of the word
    previous_digit = None

    # Map each character in the word to its Soundex code
    for char in word[1:]:
        for key, value in soundex_mapping.items():
            if char in key:
                current_digit = value
                if current_digit != previous_digit:
                    soundex_code += current_digit
                previous_digit = current_digit

    # Return the Soundex code with 4 characters, padding with zeros if necessary
    return soundex_code[:4].ljust(4, '0')

# Step 2: Levenshtein distance for spelling correction
def levenshtein_distance(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                cost = 0
            else:
                cost = 1
            dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + cost)

    return dp[m][n]

# Step 3: N-grams and Jaccard Coefficient for similarity
def ngrams(word, n):
    return [word[i:i + n] for i in range(len(word) - n + 1)]

def jaccard_coefficient(word1, word2, n=2):
    ngrams1 = set(ngrams(word1, n))
    ngrams2 = set(ngrams(word2, n))
    intersection = len(ngrams1 & ngrams2)
    union = len(ngrams1 | ngrams2)
    return intersection / union

# Step 4: Permuterm Index for wildcard handling
def create_permuterm_index(dictionary):
    permuterm_index = defaultdict(list)
    for word in dictionary:
        for i in range(len(word)):
            rotated = word[i:] + '$' + word[:i]
            permuterm_index[rotated].append(word)
    return permuterm_index

def search_wildcard(query, permuterm_index):
    if query.endswith("*"):
        query_base = query[:-1]  # Remove '*' for the base word
        results = []

        # Search for any keys in permuterm_index that contain the query as a prefix
        for term in permuterm_index.keys():
            if query_base in term:
                results.extend(permuterm_index[term])
        
        # Return unique results (as a set) to avoid duplicates
        return list(set(results))
    else:
        return []

# Step 5: Query processing function integrating Soundex, Levenshtein, and Wildcard search
def query_processing(query, inverted_index, dictionary, permuterm_index):
    # 1. Check for exact matches in the inverted index
    exact_matches = inverted_index.get(query, [])

    # 2. Perform Soundex-based phonetic matching
    soundex_code = soundex(query)
    phonetic_matches = []
    for term in dictionary:
        if soundex(term) == soundex_code:
            phonetic_matches.extend(inverted_index.get(term, []))

    # 3. Perform spelling correction using Levenshtein distance
    spelling_corrections = []
    for term in dictionary:
        lev_distance = levenshtein_distance(query, term)
        if lev_distance <= 2:  # Consider words with a small Levenshtein distance
            spelling_corrections.extend(inverted_index.get(term, []))

    # 4. Perform wildcard search using the permuterm index
    wildcard_matches = search_wildcard(query, permuterm_index)

    return {
        "exact_matches": exact_matches,
        "phonetic_matches": phonetic_matches,
        "spelling_corrections": spelling_corrections,
        "wildcard_matches": wildcard_matches
    }
dictionary = list(inverted_index.keys())

# Now build the permuterm index
permuterm_index = create_permuterm_index(dictionary)

query = "comp*"  # Example wildcard query
results = query_processing(query, inverted_index, dictionary, permuterm_index)
# Example usage
# Assuming inverted_index and dictionary are already defined
permuterm_index = create_permuterm_index(dictionary)

query = "comp*"  # Example wildcard query
results = query_processing(query, inverted_index, dictionary, permuterm_index)

# Output the results
print(f"Results for the query '{query}':")
print("Exact Matches:", results['exact_matches'])
print("Phonetic Matches:", results['phonetic_matches'])
print("Spelling Corrections:", results['spelling_corrections'])
print("Wildcard Matches:", results['wildcard_matches'])
```

```python
class TreeNode:
    def __init__(self, key, value):
        self.left = None
        self.right = None
        self.key = key
        self.value = value

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, key, value):
        if not self.root:
            self.root = TreeNode(key, value)
        else:
            self._insert_recursive(self.root, key, value)

    def _insert_recursive(self, node, key, value):
        if key < node.key:
            if node.left is None:
                node.left = TreeNode(key, value)
            else:
                self._insert_recursive(node.left, key, value)
        elif key > node.key:
            if node.right is None:
                node.right = TreeNode(key, value)
            else:
                self._insert_recursive(node.right, key, value)
        else:  # If the key is already present, update the value
            node.value = value

    def search(self, key):
        return self._search_recursive(self.root, key)

    def _search_recursive(self, node, key):
        if node is None:
            return None
        if key == node.key:
            return node.value
        elif key < node.key:
            return self._search_recursive(node.left, key)
        else:
            return self._search_recursive(node.right, key)

    def inorder_traversal(self):
        return self._inorder_recursive(self.root)

    def _inorder_recursive(self, node):
        result = []
        if node:
            result = self._inorder_recursive(node.left)
            result.append((node.key, node.value))
            result += self._inorder_recursive(node.right)
        return result

# Example inverted index with document IDs
inverted_index = {
    "august": ['GX253-43-11798479-701', 'GX262-86-10646381-701'],
    "member": ['GX104-28-6788626-701', 'GX270-62-10329938-701'],
    "cat": ['GX001-90-1790440-701'],
    "dog": ['GX012-25-15935953-703'],
}

# Insert the inverted index into the binary tree
binary_tree = BinaryTree()
for key, value in inverted_index.items():
    binary_tree.insert(key, value)

# Searching for a posting list
search_key = "cat"
value = binary_tree.search(search_key)
if value is not None:
    print(f"Posting list for '{search_key}': {value}")
else:
    print(f"'{search_key}' not found in the binary tree.")

# Inorder traversal
print("Inorder Traversal of Binary Tree:", binary_tree.inorder_traversal())

# Heap Implementation
import heapq

class HeapDictionary:
    def __init__(self):
        self.heap = []
        self.dictionary = {}

    def insert(self, key, value):
        if key in self.dictionary:
            # Update existing key
            self.remove(key)
        self.dictionary[key] = value
        heapq.heappush(self.heap, (key, value))

    def remove(self, key):
        if key in self.dictionary:
            del self.dictionary[key]
            # Rebuild the heap
            self.heap = [(k, v) for k, v in self.heap if k != key]
            heapq.heapify(self.heap)

    def search(self, key):
        return self.dictionary.get(key, None)

    def get_min(self):
        return self.heap[0] if self.heap else None

    def pop_min(self):
        if self.heap:
            key, value = heapq.heappop(self.heap)
            del self.dictionary[key]
            return key, value
        return None

# Insert the inverted index into the heap
heap_dict = HeapDictionary()
for key, value in inverted_index.items():
    heap_dict.insert(key, value)

# Searching for a value
search_key = "dog"
value = heap_dict.search(search_key)
if value is not None:
    print(f"Posting list for '{search_key}': {value}")
else:
    print(f"'{search_key}' not found in the heap.")

# Getting the minimum key-value pair
min_pair = heap_dict.get_min()
print("Minimum key-value pair in heap:", min_pair)

# Popping the minimum key-value pair
popped_pair = heap_dict.pop_min()
print("Popped minimum pair:", popped_pair)
print("Heap after popping:", heap_dict.heap)
```

```python
import time
import json
from collections import defaultdict
import os
import re
import logging

# Sample corpus file path (replace with actual path)
corpus_path = 'path_to_your_reduced_file'  # Update this with your actual dataset file path

# Setup logging
logging.basicConfig(level=logging.INFO)

# Timer decorator for measuring time
def time_it(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Time taken by {func.__name__}: {end_time - start_time:.4f} seconds")
        return result
    return wrapper

# Function to parse the custom document format
def parse_custom_corpus(file_path):
    documents = []
    doc_id = None
    doc_text = []
    
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                if "<DOCNO>" in line:
                    doc_id = line.strip().replace("<DOCNO>", "").replace("</DOCNO>", "")
                elif "<SENTENCE>" in line:
                    sentence = line.strip().replace("<SENTENCE>", "").replace("</SENTENCE>", "")
                    doc_text.append(sentence)
                elif "</DOC>" in line and doc_id:
                    documents.append({
                        "_id": doc_id,
                        "text": ' '.join(doc_text)
                    })
                    doc_id = None
                    doc_text = []
    except FileNotFoundError:
        print(f"Error: File {file_path} not found.")
        return None
    
    return documents

# Basic tokenization with stopword removal and regex
def tokenize(text, stopwords):
    terms = re.findall(r'\b\w+\b', text.lower())
    return [term for term in terms if term not in stopwords]

# BSBI Blocked Sort-Based Indexing Algorithm
@time_it
def bsbi_index_construction(corpus_file, block_size=10000):
    blocks = []
    block_num = 0
    stopwords = set(['the', 'is', 'in', 'and', 'of', 'to'])  # Simplified stopword list

    try:
        documents = parse_custom_corpus(corpus_file)
        if not documents:
            return

        current_block = []
        for doc in documents:
            current_block.append(doc)

            if len(current_block) >= block_size:
                block_num += 1
                logging.info(f"Processing BSBI block {block_num}")
                bsbi_invert(current_block, f'block_{block_num}.txt', stopwords)
                blocks.append(f'block_{block_num}.txt')
                current_block = []

        if current_block:  # process remaining documents
            block_num += 1
            logging.info(f"Processing BSBI block {block_num}")
            bsbi_invert(current_block, f'block_{block_num}.txt', stopwords)
            blocks.append(f'block_{block_num}.txt')

        merged_index = merge_blocks(blocks)
        return merged_index

    except Exception as e:
        print(f"An error occurred during BSBI index construction: {str(e)}")
        return None

# BSBI-Invert: Invert and write to disk
@time_it
def bsbi_invert(block, output_file, stopwords):
    inverted_index = defaultdict(list)

    for doc in block:
        doc_id = doc['_id']
        terms = tokenize(doc['text'], stopwords)
        term_frequencies = defaultdict(int)
        for term in terms:
            term_frequencies[term] += 1
        
        for term, freq in term_frequencies.items():
            inverted_index[term].append((doc_id, freq))

    sorted_terms = sorted(inverted_index.items())
    with open(output_file, 'w') as f:
        for term, postings in sorted_terms:
            f.write(f"{term}: {postings}\n")

# Merging the blocks for BSBI and SPIMI
@time_it
def merge_blocks(blocks):
    final_index = defaultdict(list)

    for block_file in blocks:
        try:
            with open(block_file, 'r') as f:
                for line in f:
                    term, postings = line.strip().split(': ')
                    postings = eval(postings)
                    final_index[term].extend(postings)
        except FileNotFoundError:
            print(f"Error: Block file {block_file} not found.")
            continue

    return final_index

# SPIMI Algorithm: Single-Pass In-Memory Indexing
@time_it
def spimi_invert(corpus_file, block_size=10000):
    block_num = 0
    blocks = []
    stopwords = set(['the', 'is', 'in', 'and', 'of', 'to'])  # Simplified stopword list

    try:
        documents = parse_custom_corpus(corpus_file)
        if not documents:
            return None

        current_block = []
        for doc in documents:
            current_block.append(doc)

            if len(current_block) >= block_size:
                block_num += 1
                logging.info(f"Processing SPIMI block {block_num}")
                output_file = f'spimi_block_{block_num}.txt'
                spimi_block_invert(current_block, output_file, stopwords)
                blocks.append(output_file)
                current_block = []

        if current_block:
            block_num += 1
            logging.info(f"Processing SPIMI block {block_num}")
            output_file = f'spimi_block_{block_num}.txt'
            spimi_block_invert(current_block, output_file, stopwords)
            blocks.append(output_file)

        merged_index = merge_blocks(blocks)
        return merged_index

    except Exception as e:
        print(f"An error occurred during SPIMI index construction: {str(e)}")
        return None

@time_it
def spimi_block_invert(block, output_file, stopwords):
    dictionary = defaultdict(list)

    for doc in block:
        doc_id = doc['_id']
        terms = tokenize(doc['text'], stopwords)
        term_frequencies = defaultdict(int)
        for term in terms:
            term_frequencies[term] += 1
        
        for term, freq in term_frequencies.items():
            dictionary[term].append((doc_id, freq))

    sorted_terms = sorted(dictionary.items())
    with open(output_file, 'w') as f:
        for term, postings in sorted_terms:
            f.write(f"{term}: {postings}\n")

# Utility to visualize the inverted index
def visualize_index(index):
    for term, doc_ids in list(index.items())[:10]:  # Limit to 10 terms for simplicity
        print(f"Term: {term}, Document IDs: {doc_ids}")

# Main Execution: BSBI vs SPIMI Comparison
def main():
    print("Starting BSBI Index Construction")
    bsbi_index = bsbi_index_construction(corpus_path)

    print("\nStarting SPIMI Index Construction")
    spimi_index = spimi_invert(corpus_path)

    # Visualizing indexes
    print("\nBSBI Inverted Index (first 10 terms):")
    if bsbi_index:
        visualize_index(bsbi_index)

    print("\nSPIMI Inverted Index (first 10 terms):")
    if spimi_index:
        visualize_index(spimi_index)

if __name__ == "__main__":
    main()
```

```python
import time
import re
import math
from collections import defaultdict
import matplotlib.pyplot as plt

# Function to convert document IDs to integers by extracting the numeric part
def convert_to_integers(postings_list):
    try:
        return [int(re.search(r'GX\d{3}-\d{2}-(\d+)-\d{3}', doc_id).group(1)) for doc_id in postings_list]
    except AttributeError:
        raise ValueError("Unable to extract numeric document IDs. Ensure the format matches.")

# Gap encode function
def gap_encode(postings_list):
    gaps = [postings_list[0]]  # First posting remains unchanged
    for i in range(1, len(postings_list)):
        gaps.append(postings_list[i] - postings_list[i - 1])
    return gaps

# Variable-byte encoding function
def variable_byte_encode(number):
    bytes_ = []
    while True:
        bytes_.insert(0, number % 128)  # Insert at the front of the list
        if number < 128:
            break
        number //= 128
    bytes_[-1] += 128  # Mark the last byte with a continuation bit
    return bytes_

# Gamma encoding function
def gamma_encode(number):
    if number == 0:
        return [0]  # Encode 0 as a single 0
    length = number.bit_length()
    unary = [1] * (length - 1) + [0]
    binary = bin(number)[2:]
    return unary + [int(bit) for bit in binary]

# Dictionary compression using blocking
def blocking_compression(dictionary, k=4):
    sorted_terms = sorted(dictionary.keys())
    compressed_dict = []
    for i in range(0, len(sorted_terms), k):
        block = sorted_terms[i:i+k]
        # Store the block and its length
        compressed_dict.append((len(block), block))
    return compressed_dict

# Compressing the inverted index
def compress_inverted_index(inverted_index):
    compressed_index = {}
    
    # Dictionary compression
    compressed_dict = blocking_compression(inverted_index)
    print(f"Compressed Dictionary Size (Blocking): {sum(len(term) for block in compressed_dict for term in block[1])} bytes")
    
    for term, postings_list in inverted_index.items():
        postings_list = convert_to_integers(postings_list)  # Convert to integers by extracting numeric part
        postings_list = sorted(postings_list)  # Sort the list of document IDs
        
        # Apply gap encoding
        gaps = gap_encode(postings_list)
        
        # Apply variable-byte encoding
        encoded_gaps = [variable_byte_encode(gap) for gap in gaps]
        
        compressed_index[term] = encoded_gaps

    return compressed_index, compressed_dict

# Decompressing the inverted index (for reference)
def decompress_inverted_index(compressed_index):
    decompressed_index = {}
    for term, encoded_gaps in compressed_index.items():
        gaps = []
        
        # Decode variable-byte encoded gaps
        for encoded_gap in encoded_gaps:
            gap = 0
            for byte in encoded_gap:
                if byte >= 128:
                    gap = gap * 128 + (byte - 128)
                    break
                gap = gap * 128 + byte
            gaps.append(gap)
        
        # Convert gaps back to postings list
        postings_list = [gaps[0]]
        for i in range(1, len(gaps)):
            postings_list.append(postings_list[-1] + gaps[i])
        
        decompressed_index[term] = postings_list
        
    return decompressed_index

# Statistical Analysis for Heaps' Law
def analyze_vocabulary(collection_size, terms_count):
    k = 100  # Sample value for k
    b = 0.5  # Sample value for b
    predicted_vocabulary_size = k * (collection_size ** b)
    print(f"Predicted Vocabulary Size (Heaps' Law): {predicted_vocabulary_size:.2f}")
    return predicted_vocabulary_size

# Zipf's Law Analysis
def analyze_zipf(terms_frequency):
    zipf_distribution = {}
    for i, freq in enumerate(sorted(terms_frequency, reverse=True), start=1):
        zipf_distribution[i] = freq
    return zipf_distribution

# Evaluating compression performance
def evaluate_compression(original_size, compressed_size):
    space_saved = original_size - compressed_size
    compression_ratio = compressed_size / original_size
    print(f"Space saved: {space_saved} bytes")
    print(f"Compression ratio: {compression_ratio:.2f}")

# Visualization Functions
def plot_zipf_distribution(zipf_distribution):
    ranks = list(zipf_distribution.keys())
    frequencies = list(zipf_distribution.values())
    
    plt.figure(figsize=(10, 5))
    plt.plot(ranks, frequencies, marker='o')
    plt.yscale('log')
    plt.xscale('log')
    plt.title("Zipf's Law Distribution")
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

def plot_heap_law(predicted_vocabulary_size, actual_terms_count):
    plt.figure(figsize=(10, 5))
    plt.bar(['Predicted Vocabulary Size', 'Actual Vocabulary Size'], 
            [predicted_vocabulary_size, actual_terms_count], color=['blue', 'orange'])
    plt.title("Heaps' Law: Predicted vs Actual Vocabulary Size")
    plt.ylabel('Vocabulary Size')
    plt.yscale('log')  # Logarithmic scale for better visualization
    plt.show()


def plot_vocabulary_vs_total_words(vocabulary_size, total_words):
    plt.figure(figsize=(10, 5))
    plt.bar(['Vocabulary Size', 'Total Words'], [vocabulary_size, total_words], color=['green', 'red'])
    plt.title("Vocabulary Size vs Total Number of Words")
    plt.ylabel('Count')
    plt.show()

# Running the compression process
def run_compression_process(inverted_index):
    print("Compressing the inverted index...")
    start_time = time.time()
    compressed_index, compressed_dict = compress_inverted_index(inverted_index)
    compression_time = time.time() - start_time
    print(f"Inverted index compressed in {compression_time:.4f} seconds.")
    
    print("Decompressing the inverted index...")
    start_time = time.time()
    decompressed_index = decompress_inverted_index(compressed_index)
    decompression_time = time.time() - start_time
    print(f"Inverted index decompressed in {decompression_time:.4f} seconds.")
    
    # Evaluate compression performance
    original_size = sum(len(doc) for postings in inverted_index.values() for doc in postings)
    compressed_size = sum(len(gap) for encoded in compressed_index.values() for gap in encoded)
    evaluate_compression(original_size, compressed_size)
    
    # Analyze vocabulary size using Heaps' Law
    total_tokens = sum(len(postings) for postings in inverted_index.values())
    actual_vocabulary_size = len(inverted_index)  # Unique terms in inverted index
    predicted_size = analyze_vocabulary(total_tokens, actual_vocabulary_size)

    # Plotting results
    plot_heap_law(predicted_size, actual_vocabulary_size)
    plot_vocabulary_vs_total_words(actual_vocabulary_size, total_tokens)
    
    return compressed_index, decompressed_index, compressed_dict

# Example usage with your dataset:
inverted_index = build_inverted_index(parsed_documents)

# Run the compression and decompression process
compressed_index, decompressed_index, compressed_dict = run_compression_process(inverted_index)

# Optional: Displaying results
print("\nCompressed Index (Sample):")
for term, encoded in list(compressed_index.items())[:2]:
    print(f"Term: {term}, Encoded Gaps: {encoded}")

print("\nDecompressed Index (Sample):")
for term, postings in list(decompressed_index.items())[:2]:
    print(f"Term: {term}, Postings List: {postings}")

# Zipf's Law Analysis Example
terms_frequency = [3, 2, 1, 1]  # Sample frequencies for Zipf's Law
zipf_distribution = analyze_zipf(terms_frequency)
print("\nZipf's Law Distribution:")
for rank, freq in zipf_distribution.items():
    print(f"Rank: {rank}, Frequency: {freq}")

# Visualizing Zipf's Law distribution
plot_zipf_distribution(zipf_distribution)
```

# Term Frequency-Inverse Document Frequency (TF-IDF)
# Logarithmic Term Frequency
# Vector Normalization
# Cosine Similarity
# Jaccard Similarity Coefficient
# Cumulative TF-IDF Score Ranking

```python
import math
from collections import defaultdict

# Function to load documents from a file, limiting to the first 20 documents
def load_documents(file_path):
    documents = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for i, line in enumerate(file):
            if i >= 20:  # Limit to the first 20 documents
                break
            # Assuming each line is a document and splitting into words
            documents.append(line.strip().lower().split())
    return documents

# Function to compute term frequency (tf)
def compute_tf(document):
    tf_dict = {}
    total_terms = len(document)
    for term in document:
        tf_dict[term] = tf_dict.get(term, 0) + 1
    # Convert counts to frequency
    for term in tf_dict:
        tf_dict[term] = tf_dict[term] / total_terms
    return tf_dict

# Function to compute inverse document frequency (idf)
def compute_idf(documents):
    idf_dict = {}
    total_documents = len(documents)
    
    # Count the number of documents containing each term
    for document in documents:
        for term in set(document):
            idf_dict[term] = idf_dict.get(term, 0) + 1
    
    # Calculate idf
    for term in idf_dict:
        idf_dict[term] = math.log(total_documents / (1 + idf_dict[term]))  # Add 1 to avoid division by zero
    return idf_dict

# Function to compute log-frequency weighting
def log_frequency_weighting(tf):
    return {term: (math.log(count + 1) if count > 0 else 0) for term, count in tf.items()}

# Function to compute TF-IDF
def compute_tfidf(documents):
    tfidf_matrix = []
    idf_dict = compute_idf(documents)
    
    for document in documents:
        tf = compute_tf(document)
        tf_log_weighted = log_frequency_weighting(tf)  # Apply log-frequency weighting
        tfidf_vector = {}
        for term, tf_value in tf_log_weighted.items():
            tfidf_vector[term] = tf_value * idf_dict.get(term, 0)
        # Normalize the TF-IDF vector
        length_norm = math.sqrt(sum(weight ** 2 for weight in tfidf_vector.values()))
        if length_norm > 0:
            for term in tfidf_vector:
                tfidf_vector[term] /= length_norm  # Length normalization
        tfidf_matrix.append(tfidf_vector)
    
    return tfidf_matrix

# Function to compute cosine similarity between two TF-IDF vectors
def cosine_similarity(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])
    
    sum1 = sum([vec1[x]**2 for x in vec1.keys()])
    sum2 = sum([vec2[x]**2 for x in vec2.keys()])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)
    
    if denominator == 0:
        return 0.0
    else:
        return numerator / denominator

# Function to compute Jaccard coefficient
def jaccard_coefficient(doc1, doc2):
    set1 = set(doc1)
    set2 = set(doc2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0.0

# Main code execution
file_path = 'path_to_your_reduced_file'  # Replace with your actual file path

# Load the first 20 documents from the specified file
documents = load_documents(file_path)

# Compute TF-IDF for the documents
tfidf_results = compute_tfidf(documents)

# Display the TF-IDF results and top 20 terms for the first 20 documents
print("TF-IDF Results and Top 20 Terms for Each Document:")
for i, tfidf in enumerate(tfidf_results):
    # Sort terms by their TF-IDF weight
    sorted_terms = sorted(tfidf.items(), key=lambda x: x[1], reverse=True)[:20]
    print(f"\nDocument {i + 1}:")
    for term, weight in sorted_terms:
        print(f"Term: {term}, TF-IDF Weight: {weight:.4f}")

# Compute cosine similarity between the first two documents if available
if len(tfidf_results) > 1:
    similarity = cosine_similarity(tfidf_results[0], tfidf_results[1])
    print(f"\nCosine Similarity between Document 1 and Document 2: {similarity:.4f}")

# Compute Jaccard coefficient between the first two documents if available
if len(documents) > 1:
    jaccard_sim = jaccard_coefficient(documents[0], documents[1])
    print(f"\nJaccard Similarity between Document 1 and Document 2: {jaccard_sim:.4f}")

# Ranking the documents based on their scores
cumulative_scores = [(i, sum(tfidf.values())) for i, tfidf in enumerate(tfidf_results)]
# Sort by cumulative score and select the top documents
top_documents = sorted(cumulative_scores, key=lambda x: x[1], reverse=True)[:20]

print("\nTop Documents Based on Cumulative TF-IDF Scores:")
for doc_index, score in top_documents:
    print(f"Document {doc_index + 1}: Score = {score:.4f}")
```
